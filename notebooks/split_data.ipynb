```python
# Étape 4 — Séparation des données (Training / Test)

# Importer les librairies nécessaires
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split


# Charger le dataset
data = pd.read_csv("../sonar.all-data.csv", header=None)

print("Dataset chargé avec succès")
print("Shape du dataset :", data.shape)


# Séparer les données (X) et les labels (Y)

# X = données sonar (colonnes 0 à 59)
X = data.iloc[:, 0:60]

# Y = résultat (colonne 60)
Y = data.iloc[:, 60]

print("\nDimensions avant encodage")
print("X shape :", X.shape)
print("Y shape :", Y.shape)


# Encoder les labels
# Transformer R → 0 et M → 1

encoder = LabelEncoder()
Y = encoder.fit_transform(Y)

print("\nAperçu des labels encodés :")
print(Y[:10])


# Séparation des données en Training et Test
# 80% entraînement
# 20% test

X_train, X_test, Y_train, Y_test = train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)

print("\nSéparation effectuée avec succès !")


# Vérifier les dimensions après séparation

print("\nDimensions après séparation :")

print("Taille X_train :", X_train.shape)
print("Taille X_test  :", X_test.shape)
print("Taille Y_train :", Y_train.shape)
print("Taille Y_test  :", Y_test.shape)


# Vérifier la répartition des classes

unique_train, counts_train = np.unique(Y_train, return_counts=True)
unique_test, counts_test = np.unique(Y_test, return_counts=True)

print("\nRépartition dans Y_train :")
for label, count in zip(unique_train, counts_train):
    nom = encoder.inverse_transform([label])[0]
    print(f"{nom} ({label}) → {count} exemples")

print("\nRépartition dans Y_test :")
for label, count in zip(unique_test, counts_test):
    nom = encoder.inverse_transform([label])[0]
    print(f"{nom} ({label}) → {count} exemples")


# Résumé final

total = X_train.shape[0] + X_test.shape[0]

print("\n=== RÉSUMÉ ===")
print(f"Total des données : {total}")
print(f"Données d'entraînement : {X_train.shape[0]} ({X_train.shape[0]/total*100:.0f}%)")
print(f"Données de test : {X_test.shape[0]} ({X_test.shape[0]/total*100:.0f}%)")

print("\nLes données sont prêtes pour l'étape 5 : entraînement du modèle.")
```
