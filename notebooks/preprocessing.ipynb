```python
# preprocessing.ipynb
# Étape 3 : Préparation des données

# Importer les librairies nécessaires
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# 2Charger le dataset
# Chaque notebook recharge les données pour être exécuté indépendamment
data = pd.read_csv("../sonar.all-data.csv", header=None)

print("Dataset chargé avec succès")
print("Shape du dataset :", data.shape)


# 3Vérifier les valeurs manquantes
print("\nVérification des valeurs manquantes :")
print(data.isnull().sum())


# Séparer les données et les labels

# X = données sonar (colonnes 0 à 59)
X = data.iloc[:, 0:60]

# Y = résultat (colonne 60)
Y = data.iloc[:, 60]

print("\nAperçu des données X (features) :")
print(X.head())

print("\nAperçu des labels Y :")
print(Y.head())


# Vérifier la répartition des classes
print("\nRépartition des classes avant encodage :")
print(Y.value_counts())


# Encoder les labels
# Transformer R → 0 et M → 1

encoder = LabelEncoder()
Y = encoder.fit_transform(Y)

print("\nLabels après transformation :")
print(Y[:10])


# Vérification finale

print("\nDimensions finales :")
print("X shape :", X.shape)
print("Y shape :", Y.shape)


print("\nEncodage terminé avec succès.")
```
